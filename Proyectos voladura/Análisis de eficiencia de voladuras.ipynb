## Análisis de eficiencia de voladuras

Script para calcular y evaluar la eficiencia de las voladuras, considerando parámetros como la distribución del tamaño de fragmentos, la energía consumida y el consumo específico de explosivos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_eficiencia(distribucion_tamanios, energia_consumida, consumo_explosivos):
    """Calcula la eficiencia de la voladura."""
    tamano_medio = np.mean(distribucion_tamanios)
    tamano_maximo = np.max(distribucion_tamanios)
    tamano_minimo = np.min(distribucion_tamanios)
    energia_especifica = energia_consumida / tamano_medio
    consumo_explosivos_especifico = consumo_explosivos / tamano_medio
    eficiencia = tamano_medio / (energia_especifica * consumo_explosivos_especifico)
    return eficiencia, tamano_medio, tamano_maximo, tamano_minimo, energia_especifica, consumo_explosivos_especifico
```

```python
# Generación de datos de ejemplo
np.random.seed(0)
distribucion_tamanios = np.random.randint(1, 20, size=1000)
energia_consumida = np.random.uniform(4000, 8000, size=1)[0]
consumo_explosivos = np.random.uniform(800, 1200, size=1)[0]
```

```python
# Cálculo de la eficiencia de la voladura
eficiencia, tamano_medio, tamano_maximo, tamano_minimo, energia_especifica, consumo_explosivos_especifico = calcular_eficiencia(
    distribucion_tamanios, energia_consumida, consumo_explosivos)
```

```python
# Visualización de los resultados
print('Resultados del Análisis de Eficiencia de Voladuras')
print('-------------------------------------------------')
print('Tamaño Medio de Fragmentos: {:.2f}'.format(tamano_medio))
print('Tamaño Máximo de Fragmentos: {:.2f}'.format(tamano_maximo))
print('Tamaño Mínimo de Fragmentos: {:.2f}'.format(tamano_minimo))
print('Energía Específica: {:.2f} kJ/t'.format(energia_especifica))
print('Consumo de Explosivos Específico: {:.2f} kg/t'.format(consumo_explosivos_especifico))
print('Eficiencia de la Voladura: {:.2f}'.format(eficiencia))
```

```python
# Visualización de la distribución de tamaños de fragmentos
plt.figure(figsize=(20, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(distribucion_tamanios, bins='auto', alpha=0.75)
plt.xlabel('Tamaño de Fragmentos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Tamaños de Fragmentos')
plt.grid(True)
```

```python
# Diagrama de dispersión
plt.figure(figsize=(20, 6))

plt.subplot(1, 2, 2)
valores_x = np.arange(len(distribucion_tamanios))
plt.scatter(valores_x, distribucion_tamanios, marker='o', color='b', label='Fragmentos')
plt.plot(valores_x, [tamano_medio] * len(distribucion_tamanios), linestyle='--', color='r', label='Tamaño Medio')
plt.xlabel('Índice del Fragmento')
plt.ylabel('Tamaño de Fragmentos')
plt.title('Diagrama de Dispersión')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
